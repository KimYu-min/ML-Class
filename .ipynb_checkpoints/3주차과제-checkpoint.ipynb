```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
from sklearn.model_selection import train_test_split
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

### KNN 회귀

```python
from sklearn.neighbors import KNeighborsRegressor

X, y = mglearn.datasets.make_wave(n_samples=40)

# wave 데이터 세트를 훈련 세트와 테스트 세트로 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# 이웃의 수를 3으로 하여 모델 객체 생성
reg = KNeighborsRegressor(n_neighbors=3)

# 훈련 데이터와 타깃으로 모델을 학습
reg.fit(X_train, y_train)
```

```python
print("테스트 세트 예측 : \n{}".format(reg.predict(X_test)))
# R^2 : 결정 계수
print("테스트 세트 R^2 : {:.2f}".format(reg.score(X_test, y_test)))
```

```python
# 테스트 세트 비율을 0.4로 변경
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.4)

reg = KNeighborsRegressor(n_neighbors=3)
reg.fit(X_train, y_train)
```

```python
print("테스트 세트 예측 : \n{}".format(reg.predict(X_test)))
print("테스트 세트 R^2 : {:.2f}".format(reg.score(X_test, y_test)))
```

### 선형 회귀

```python
from sklearn.linear_model import LinearRegression

X, y = mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

lr = LinearRegression().fit(X_train, y_train)
```

```python
print("훈련 세트 점수 : {:.2f}".format(lr.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(lr.score(X_test, y_test)))
```

```python
# lr의 기울기와 y 절편
print("lr.coef_ : ", lr.coef_)
print("lr.intercept_ : ", lr.intercept_)
```

#### - 선형 회귀로 Boston 주택 가격 예측

```python
from sklearn.datasets import load_boston

boston = load_boston()
df = pd.DataFrame(boston.data, columns=boston.feature_names)

# 방 개수와 주택 가격 산점도
plt.scatter(boston.data[:, 5], boston.target, color='r', s=10)
```

```python
# 특징 하나로 선형 회귀 실험

X = boston.data[:, 5]
print(X.shape)

# LinearRegression 함수에 인자를 전달하기 위해서
# 각 원소의 값이 담긴 2차원 배열을 (n, 1) 형태로 만들어야 함
# atleast_2d : 2차원 배열 데이터를 (1, n) 형태로 변형
X_1 = np.atleast_2d(X)
X_2 = np.transpose(X_1)
print(X_2.shape)

y = boston.target
lr.fit(X_2, y)

y_predicted = lr.predict(X_2)

print("w[0] : %f   b : %f" % (lr.coef_[0], lr.intercept_))
```

```python
# 선형 회귀 모델 그리기
plt.plot(X_2, y_predicted)
plt.plot(X_2, y, 'o', color='r', markersize=3)
```

```python
# 선형 회귀 성능 측정
X_train, X_test, y_train, y_test = train_test_split(boston.data, boston.target, random_state=0)
lr = LinearRegression().fit(X_train, y_train)
print("훈련 세트 점수 : {:.2f}".format(lr.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(lr.score(X_test, y_test)))
```

#### - KNN 회귀

```python
# KNN 회귀 성능 측정
lr = KNeighborsRegressor().fit(X_train, y_train)
print("훈련 세트 점수 : {:.2f}".format(lr.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(lr.score(X_test, y_test)))
```

### 릿지 회귀

```python
from sklearn.linear_model import Ridge
# default = 1.0

# alpha가 클수록 계수 편차가 적음 (과소적합)
# alpha가 작을수록 계수 편차가 큼 (과대적합)
# 테스트 점수 향상
ridge = Ridge().fit(X_train, y_train)
print("훈련 세트 점수 : {:.2f}".format(ridge.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(ridge.score(X_test, y_test)))
```

```python
# alpha가 10일 때 : 과소적합
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
print("훈련 세트 점수 : {:.2f}".format(ridge10.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(ridge10.score(X_test, y_test)))
```

### 라쏘 회귀

```python
from sklearn.linear_model import Lasso

# alpha = 1 : 적은 특성 사용, 낮은 성능, 과소적합
lasso = Lasso().fit(X_train, y_train)
print("훈련 세트 점수 : {:.2f}".format(lasso.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(lasso.score(X_test, y_test)))
print("사용한 특성의 수 : {}".format(np.sum(lasso.coef_ != 0)))
```

```python
# alpha = 0.01 : 패널티 효과 감소, 복잡도 증가
lasso001 = Lasso(alpha=0.01, max_iter=100000).fit(X_train, y_train)
print("훈련 세트 점수 : {:.2f}".format(lasso001.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(lasso001.score(X_test, y_test)))
print("사용한 특성의 수 : {}".format(np.sum(lasso001.coef_ != 0)))
```

```python
# alpha를 아주 작게 만들면 선형 모델과 유사
lasso00001 = Lasso(alpha=0.0001, max_iter=100000).fit(X_train, y_train)
print("훈련 세트 점수 : {:.2f}".format(lasso00001.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(lasso00001.score(X_test, y_test)))
print("사용한 특성의 수 : {}".format(np.sum(lasso00001.coef_ != 0)))
```

```python
plt.plot(lasso.coef_, 's', label="Lasso alpha = 1")
plt.plot(lasso001.coef_, '^', label="Lasso alpha = 0.01")
plt.plot(lasso00001.coef_, 'v', label="Lasso alpha = 0.0001")

# alpha = 1 : 계수 대부분이 0, 나머지도 크기가 작음
# alpha = 0.01 : 많은 특성이 0
# alpha = 0.0001 : 값이 커지고 규제를 받지 않음
plt.plot(ridge10.coef_, 'o', label="Ridge alpha = 0.1")
plt.legend(ncol=2, loc=(0, 1.05))
plt.ylim(-25, 25)
plt.xlabel("계수 목록")
plt.ylabel("계수 크기")
```

### 로지스틱 회귀

```python
from sklearn.linear_model import LogisticRegression
from sklearn import datasets

iris = datasets.load_iris()

print(list(iris.keys()))
print("타깃의 이름 : {}".format(iris['target_names']))
print("특성의 이름 : {}".format(iris['feature_names']))

X = iris["data"][:,3:]  # petal width
print(X.shape)
y = (iris["target"]==2).astype(np.int)  # virginica면 1, 아니면 0
print(y.shape)
```

```python
# penalty = l2 (default)
log_reg = LogisticRegression()
log_reg.fit(X, y)

# 꽃잎의 너비가 0 ~ 3cm인 꽃에 대한 모델의 추정 확률
X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)

# 출력 값은 0 또는 1
plt.plot(X, y, "b.")
plt.plot(X_new, y_proba[:,1], "g-", label="Iris-Virginica")
plt.plot(X_new, y_proba[:,0], "b--", label="Not Iris-Virginica")
plt.xlabel("Petal Width", fontsize=14)
plt.ylabel("Probability", fontsize=14)
plt.legend(loc="upper left", fontsize=14)
plt.show()
```

### 로지스틱 회귀로 유방암 데이터 예측

```python
from sklearn.datasets import load_breast_cancer
# sample 수는 569, feature 수는 30

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target,
                                                    stratify=cancer.target,
                                                    random_state=42)

logreg = LogisticRegression(max_iter=5000).fit(X_train, y_train)
print("훈련 세트 점수 : {:.3f}".format(logreg.score(X_train, y_train)))
print("테스트 세트 점수 : {:.3f}".format(logreg.score(X_test, y_test)))
```

```python
# C : 규제
# 1/100 : 패널티가 작아져서 규제 완화, 훈련 세트에 맞추려 함 (과적합)
logreg100 = LogisticRegression(C=100, max_iter=5000).fit(X_train, y_train)
print("훈련 세트 점수 : {:.3f}".format(logreg100.score(X_train, y_train)))
print("테스트 세트 점수 : {:.3f}".format(logreg100.score(X_test, y_test)))
```

```python
# 100 : 패널티가 커져서 규제 강화, 다수의 포인트에 맞추려 함 (일반화)
logreg001 = LogisticRegression(C=0.01, max_iter=5000).fit(X_train, y_train)
print("훈련 세트 점수 : {:.3f}".format(logreg001.score(X_train, y_train)))
print("테스트 세트 점수 : {:.3f}".format(logreg001.score(X_test, y_test)))
```

```python
# l1 : 계수 크기가 0인 경우가 많음
for C, marker in zip([0.001, 1, 100], ['o', '^', 'v']):
    lr_l1 = LogisticRegression(solver='liblinear', C=C, penalty="l1",
                              max_iter=1000).fit(X_train, y_train)
    print("C = {:.3f} l1 로지스틱 회귀의 훈련 정확도 : {:.2f}".format(
    C, lr_l1.score(X_train, y_train)))
    print("C = {:.3f} l1 로지스틱 회귀의 테스트 정확도 : {:.2f}".format(
    C, lr_l1.score(X_test, y_test)))
    plt.plot(lr_l1.coef_.T,marker, label="C = {:.3f}".format(C))
    
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.xlabel("특성")
plt.ylabel("계수 크기")
plt.ylim(-5, 5)
plt.legend(loc=3)
```

```python
# l2 : 계수 크기가 큰 경우가 많음
for C, marker in zip([0.001, 1, 100], ['o', '^', 'v']):
    lr_l2 = LogisticRegression(solver='liblinear', C=C, penalty="l2",
                              max_iter=1000).fit(X_train, y_train)
    print("C = {:.3f} l2 로지스틱 회귀의 훈련 정확도 : {:.2f}".format(
    C, lr_l2.score(X_train, y_train)))
    print("C = {:.3f} l2 로지스틱 회귀의 테스트 정확도 : {:.2f}".format(
    C, lr_l2.score(X_test, y_test)))
    plt.plot(lr_l2.coef_.T,marker, label="C = {:.3f}".format(C))
    
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.xlabel("특성")
plt.ylabel("계수 크기")
plt.ylim(-5, 5)
plt.legend(loc=3)
```
