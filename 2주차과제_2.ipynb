### - KNN

```python
 # 데이터 준비
X, y = mglearn.datasets.make_forge()
```

```python
# 데이터 형태 보기
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["클래스0", "클래스1"], loc=4)
plt.xlabel("first feature")
plt.ylabel("second feature")
print("X.shape() : {}".format(X.shape))
```

```python
from sklearn.model_selection import train_test_split

# 데이터 로드
X, y = mglearn.datasets.make_forge()
print("X.shape : {}".format(X.shape), "y.shape : {}".format(y.shape))
```

```python
# 훈련 데이터와 테스트 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
print("X_train.shape : {}".format(X_train.shape), "\nX_test.shape:{}".format(X_test.shape))
```

```python
from sklearn.neighbors import KNeighborsClassifier

# 모델 선택, k = 3
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
# 훈련 데이터 학습
clf.fit(X_train, y_train)
```

```python
# 테스트 데이터
print("테스트 세트 예측 : {}".format(clf.predict(X_test)))
```

```python
# 성능 확인
print("테스트 세트 정확도 : {:.2f}".format(clf.score(X_test, y_test)))
```

### - k 값 조정 결과
#### k = 19 : 전체 포인트를 이웃으로 설정

```python
clf = KNeighborsClassifier(n_neighbors=19)
clf.fit(X_train, y_train)
print("테스트 세트 예측 : {}".format(clf.predict(X_test)))
print("테스트 세트 정확도 : {:.2f}".format(clf.score(X_test, y_test)))
```

#### 전체 포인트를 이웃으로 설정하자 정확도가 대폭 하락했다
#### 이유 : 전체 포인트를 이웃으로 둬서 예측값이 다수값인 0으로 통일됐기 때문
### - 유방암 데이터로 KNN

```python
# 데이터 로드
cancer = load_breast_cancer()
print("cancer.keys() : {}".format(cancer.keys()))
```

```python
print("shape : {}".format(cancer.data.shape))
print("클래스 별 샘플 수 : \n{}".
      format({n: v for n, v in zip(cancer.target_names, np.bincount(cancer.target))}))
```

```python
print("특성 이름 : \n{}".format(cancer.feature_names))
```

```python
# 훈련 데이터와 테스트 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(cancer.data,
                                                   cancer.target,
                                                   stratify=cancer.target,
                                                   random_state=66)
training_accuracy = []
test_accuracy = []

# 1에서 10까지 n_neighbors를 적용
neighbors_settings = range(1, 11)

for n_neighbors in neighbors_settings:
    # 모델 생성
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    
    # 훈련 세트 정확도 저장
    training_accuracy.append(clf.score(X_train, y_train))
    
    # 일반화 정확도 저장
    test_accuracy.append(clf.score(X_test, y_test))

plt.plot(neighbors_settings, training_accuracy, label='훈련 정확도')
plt.plot(neighbors_settings, test_accuracy, label='테스트 정확도')
plt.ylabel("정확도")
plt.xlabel("n_neighbors")
plt.legend()
```

### - neighbors_settings의 범위 조정
#### range(1, 50)으로 조정

```python
# 훈련 데이터와 테스트 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(cancer.data,
                                                   cancer.target,
                                                   stratify=cancer.target,
                                                   random_state=66)
training_accuracy = []
test_accuracy = []

# 1에서 10까지 n_neighbors를 적용
# range(1, 11) : 1부터 11 미만
neighbors_settings = range(1, 50)

for n_neighbors in neighbors_settings:
    # 모델 생성
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    
    # 훈련 세트 정확도 저장
    training_accuracy.append(clf.score(X_train, y_train))
    
    # 일반화 정확도 저장
    test_accuracy.append(clf.score(X_test, y_test))

plt.plot(neighbors_settings, training_accuracy, label='훈련 정확도')
plt.plot(neighbors_settings, test_accuracy, label='테스트 정확도')
plt.ylabel("정확도")
plt.xlabel("n_neighbors")
plt.legend()
```

#### k 수가 증가할수록 테스트 정확도가 불안정한 것을 확인할 수 있다
#### 과적합이 발생한 것을 알 수 있다
