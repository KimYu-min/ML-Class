```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
# 데이터 불러오기
bankruptcy = pd.read_csv('bankruptcy_data.csv')
bankruptcy[:10]
```

```python
# 6819개의 샘플과 96개의 특성
bankruptcy.shape
```

```python
# 데이터 정보
bankruptcy.info()
```

```python
# X와 y 분리
X = bankruptcy.drop('Bankrupt?', axis=1)
y = bankruptcy['Bankrupt?']
```

```python
X[:10]
```

```python
y[:10]
```

### 주성분 분석(PCA)을 통한 차원 축소

```python
# 타겟과의 상관 관계 절댓값이 0.15 이상인 특성들 추출
corrmat = bankruptcy.corr()
top_corr_features = corrmat.index[abs(corrmat["Bankrupt?"])>=0.15]

# 히트맵 출력
plt.figure(figsize=(13,10))
g = sns.heatmap(bankruptcy[top_corr_features].corr(), annot=True, cmap="RdYlGn")
plt.show()
```

#### - 타겟과 특성들 사이의 상관 관계가 거의 없다.
#### - 특성들 간의 상관 관계도 그리 강하지 않다.

```python
from sklearn import preprocessing

# 평균 0, 분산 1이 되도록 X 데이터 스케일 조정
scaler = preprocessing.StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled_pd = pd.DataFrame(X_scaled)
X_scaled_pd
```

```python
from sklearn.decomposition import PCA

# 특성 수를 2개로 축소
pca = PCA(n_components=2)
pca.fit(X_scaled)
X_pca = pca.transform(X_scaled)

print("원본 데이터 : {}".format(str(X_scaled.shape)))
print("축소 데이터 : {}".format(str(X_pca.shape)))
```

```python
# 특성을 2개 추출했을 때 각 주성분의 분산 설명량
print(pca.explained_variance_ratio_)
print(sum(pca.explained_variance_ratio_))
```

```python
columns = []

for i in range(1, 3):
    columns.append('pc{}'.format(i))
    
X_pca_pd = pd.DataFrame(X_pca, columns=columns)
X_pca_pd
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(X_pca[:,0], X_pca[:,1], c=bankruptcy['Bankrupt?'], s=70, cmap='winter')
plt.title("2D")
plt.xlabel('pc1')
plt.ylabel('pc2')
plt.show()
```

```python
# 특성 수를 7개로 축소
pca = PCA(n_components=7)
pca.fit(X_scaled)
X_pca = pca.transform(X_scaled)

print("원본 데이터 : {}".format(str(X_scaled.shape)))
print("축소 데이터 : {}".format(str(X_pca.shape)))
```

```python
print(pca.explained_variance_ratio_)
print(sum(pca.explained_variance_ratio_))
```

```python
columns = []

for i in range(1, 8):
    columns.append('pc{}'.format(i))
    
X_pca_pd = pd.DataFrame(X_pca, columns=columns)
X_pca_pd
```

### 로지스틱 회귀

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# 훈련/검증 데이터와 테스트 데이터로 분리
X_trainval, X_test, y_trainval, y_test = train_test_split(X_pca, y, stratify=y, random_state=42)

print(X_trainval.shape)
print(X_test.shape)
```

```python
# 훈련/검증 데이터를 다시 훈련 데이터와 검증 데이터로 분리
X_train, X_valid, y_train, y_valid = train_test_split(X_trainval, y_trainval, stratify=y_trainval, random_state=42)

print(X_train.shape)
print(X_valid.shape)
```

```python
# C 값에 따른 로지스틱 회귀 정확도
for C in [0.001, 0.01, 0.1, 1, 10, 100]:
    logreg = LogisticRegression(C=C)
    logreg.fit(X_train, y_train)
    score = logreg.score(X_valid, y_valid)
    print("C = {} 검증 정확도 : {:.3f}".format(C, score))
```

```python
# l1 규제일 때 C 값에 따른 로지스틱 회귀 정확도
for C in [0.001, 0.01, 0.1, 1, 10, 100]:
    logreg = LogisticRegression(C=C, solver='liblinear', penalty='l1')
    logreg.fit(X_train, y_train)
    score = logreg.score(X_valid, y_valid)
    print("C = {} 검증 정확도 : {:.3f}".format(C, score))
```

```python
# C = 1
logreg = LogisticRegression()
logreg.fit(X_trainval, y_trainval)
print("로지스틱 회귀 테스트 정확도 : {:.3f}".format(logreg.score(X_test, y_test)))
```

```python
print(logreg.coef_)
```

#### - pc1과 pc5가 타겟 결정에 큰 영향을 준다.

### 랜덤 포레스트

```python
from sklearn.ensemble import RandomForestClassifier

# 먼저 차원 축소를 하지 않은 데이터로 훈련
X_trainval2, X_test2, y_trainval2, y_test2 = train_test_split(X, y, stratify=y, random_state=42)
X_train2, X_valid2, y_train2, y_valid2 = train_test_split(X_trainval2, y_trainval2, stratify=y_trainval2, random_state=42)

for n_estimators in [10, 50, 100, 200, 500]:
    forest = RandomForestClassifier(n_estimators=n_estimators, random_state=42)
    forest.fit(X_train2, y_train2)
    score = forest.score(X_valid2, y_valid2)
    print("n_estimators = {} 검증 정확도 : {:.3f}".format(n_estimators, score))
```

```python
# n_estimators = 100
forest = RandomForestClassifier(n_estimators=100, random_state=42)
forest.fit(X_train2, y_train2)
print("랜덤 포레스트 테스트 정확도 : {:.3f}".format(forest.score(X_test2, y_test2)))
```

```python
# 특성 중요도 시각화
plt.figure(figsize=(70, 100))
n_features = X_train2.shape[1]
plt.barh(np.arange(n_features), forest.feature_importances_, align="center")
plt.yticks(np.arange(n_features), X_train2.columns.values, fontsize=60)
plt.xticks(fontsize=60)
plt.ylim(-1, n_features)
```

#### - Borrowing Dependency가 가장 중요한 특성으로, 이는 대출 의존도를 뜻한다.
#### - 두 번째로 중요한 특성은 Net Income to Stockholder's Equity로, 이는 주주 자본 순이익을 의미한다. 즉, 모든 부채를 해결한 뒤 남아 있는 자산의 양이다.
#### - 다음으로 중요한 특성들은 Net Value Growth Rate, Persistent EPS in the Last Four Seasons였다. 각각 순가치 증가율과 지난 1년 간의 지속적인 EPS를 의미한다. 여기서 EPS는 Earning Per Share로, 주당 순이익이다. 

```python
# 차원 축소한 데이터로 훈련
for n_estimators in [10, 50, 100, 200, 500]:
    forest = RandomForestClassifier(n_estimators=n_estimators, random_state=42)
    forest.fit(X_train, y_train)
    score = forest.score(X_valid, y_valid)
    print("n_estimators = {} 검증 정확도 : {:.3f}".format(n_estimators, score))
```

```python
# n_estimators = 100
forest = RandomForestClassifier(n_estimators=100, random_state=42)
forest.fit(X_trainval, y_trainval)
print("랜덤 포레스트 테스트 정확도 : {:.3f}".format(forest.score(X_test, y_test)))
```

```python
plt.figure(figsize=(5, 5))
n_features = X_train.shape[1]
plt.barh(np.arange(n_features), forest.feature_importances_, align="center")
plt.yticks(np.arange(n_features), ['pc1', 'pc2', 'pc3', 'pc4', 'pc5', 'pc6', 'pc7'])
plt.ylim(-1, n_features)
```

#### - 로지스틱 회귀와 마찬가지로 pc1이 가장 큰 중요도를 갖는 특성이다.

### 서포트 벡터 머신

```python
from sklearn.svm import SVC

for gamma in [0.001, 0.01, 0.1, 1, 10, 100]:
    print('\n')
    for C in [0.001, 0.01, 0.1, 1, 10, 100]:
        svm = SVC(gamma=gamma, C=C)
        svm.fit(X_train, y_train)
        print("gamma = {}, C = {} 검증 정확도 : {:.3f}".format(gamma, C, svm.score(X_valid, y_valid)))
```

```python
# gamma = 0.1, C = 10
svm = SVC(gamma=0.1, C=10)
svm.fit(X_trainval, y_trainval)
print("서포트 벡터 머신 테스트 정확도 : {:.3f}".format(svm.score(X_test, y_test)))
```

### 성능 평가

```python
from sklearn.metrics import confusion_matrix

pred_logreg = logreg.predict(X_test)
logreg_confusion = confusion_matrix(y_test, pred_logreg)
print("로지스틱 회귀 오차 행렬 :\n{}".format(logreg_confusion))
```

```python
pred_forest = forest.predict(X_test)
forest_confusion = confusion_matrix(y_test, pred_forest)
print("랜덤 포레스트 오차 행렬 :\n{}".format(forest_confusion))
```

```python
pred_svm = svm.predict(X_test)
svm_confusion = confusion_matrix(y_test, pred_svm)
print("서포트 벡터 머신 오차 행렬 :\n{}".format(svm_confusion))
```

```python
from sklearn.metrics import f1_score
# F1-score : 불균형 이진 분류에서 정확도보다 나은 지표

print("로지스틱 회귀 F1 score: {:.3f}".format(f1_score(y_test, pred_logreg)))
print("랜덤 포레스트 F1 score : {:.3f}".format(f1_score(y_test, pred_forest)))
print("서포트 벡터 머신 F1 score : {:.3f}".format(f1_score(y_test, pred_svm)))
```

#### - 파산 함(1)이 파산 안 함(0)보다 훨씬 적은 데이터 특성 상, TP가 더 높은 로지스틱 회귀의 성능이 가장 좋다고 할 수 있다.
